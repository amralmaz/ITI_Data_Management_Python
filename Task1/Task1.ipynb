```python
import pandas as pd

columns_imported = "c:f"

energy= pd.read_excel(r"D:\ITI\26 Python\Task 1\data\Energy Indicators.xls",
                         usecols=columns_imported,
                         skiprows=18,
                         nrows=227,
                         header=None,
                         names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'],
                         na_values=['...']
                     )

# energy=energy.replace('', np.nan, regex=True, inplace=True )
# energy.isnull().sum()
energy.head(25)
```

```python
#function to change 'Energy Supply' from petajoules to gigajoules
def gigajoules(Energy):
    return float(Energy)*1000000

energy['Energy Supply']=energy['Energy Supply'].apply(gigajoules)
```

```python
#to remove the numbers from the country name
energy['Country']=energy['Country'].str.replace('\d+', '')

#to remove the unneeded brackets and its contents 
energy['Country']=energy['Country'].str.replace(r" \(.*","")

#to transform these forms into these forms as required
energy.replace({'Country':{'Republic of Korea': 'South Korea',
'United States of America': 'United States',
'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
'China, Hong Kong Special Administrative Region': 'Hong Kong'}}, inplace=True)
```
